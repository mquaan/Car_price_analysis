```python
import pandas as pd
```

```python
car_df = pd.read_csv('car_detail.csv')
seller_df = pd.read_csv('seller_en.csv')
```

```python
car_df = car_df[['Mã tin','Xuất xứ','Tình trạng','Dòng xe','Số Km đã đi','Màu ngoại thất','Màu nội thất','Số cửa','Số chỗ ngồi','Động cơ','Hệ thống nạp nhiên liệu','Hộp số','Dẫn động','Tiêu thụ nhiên liệu','Hãng','Grade','Năm sản xuất','Tên xe','Giá',]]
seller_df = seller_df[['ad_id','Name','Address']]
```

```python
df = pd.merge(car_df, seller_df, left_on='Mã tin', right_on='ad_id', how='left')

df
```

```python
df['Năm sản xuất'] = df['Năm sản xuất'].fillna(0).astype(int).astype(object)


df['Năm sản xuất']
```

```python
df['Số Km đã đi'] = df['Số Km đã đi'].str.replace(r"[\,Km]", "", regex=True)

df['Số Km đã đi'] = df['Số Km đã đi'].astype('int64')

df['Số Km đã đi']
```

```python
df['Số cửa'] = df['Số cửa'].str.replace(r"cửa", "", regex=True)

df['Số cửa'] = df['Số cửa'].astype(int)

df['Số cửa']
```

```python
df['Số chỗ ngồi'] = df['Số chỗ ngồi'].str.replace(r"chỗ", "", regex=True)

df['Số chỗ ngồi'] = df['Số chỗ ngồi'].astype(int)

df['Số chỗ ngồi']
```

```python
df[['Nhiên liệu', 'Dung tích xi lanh (L)']]  = df['Động cơ'].str.split("\t", expand=True)
df.drop(columns='Động cơ', inplace=True)

df['Nhiên liệu']
```

```python
df['Dung tích xi lanh (L)'] = df['Dung tích xi lanh (L)'].str.replace(" L", "")

df['Dung tích xi lanh (L)']
```

```python
df['Tiêu thụ nhiên liệu'] = df['Tiêu thụ nhiên liệu'].str.replace("\tL/100Km", "")
df['Tiêu thụ nhiên liệu'] = df['Tiêu thụ nhiên liệu'].str.replace("L/100Km", "1")
df['Tiêu thụ nhiên liệu'] = df['Tiêu thụ nhiên liệu'].str.replace("K0", "1")
df['Tiêu thụ nhiên liệu'] = df['Tiêu thụ nhiên liệu'].str.replace(r"[lLK,\ ]", "", regex=True)
df['Tiêu thụ nhiên liệu'] = df['Tiêu thụ nhiên liệu'].mask(df['Tiêu thụ nhiên liệu'] == "", "1")


df['Tiêu thụ nhiên liệu'] = df['Tiêu thụ nhiên liệu'].astype(float)

df.rename(columns={'Tiêu thụ nhiên liệu': 'Tiêu thụ (L/100Km)'}, inplace=True)
```

```python
df.rename(columns={'Grade': 'Bản'}, inplace=True)
```

```python
def convert_to_numeric(value):
    try:
        if 'Tỷ' in value and not 'Triệu' in value:
            return int(value.replace('Tỷ', '').replace(' ', '')) * 1000
        elif 'Triệu' in value and not 'Tỷ' in value:
            return int(value.replace('Triệu', '').replace(' ', ''))
        else:
            return int(value.replace('Triệu', '').replace('Tỷ', '').replace(' ', ''))
    except:
        return None
    
df['Giá'] = df['Giá'].apply(convert_to_numeric)

df['Giá']
```

```python
df['Màu ngoại thất'] = df['Màu ngoại thất'].where(df['Màu ngoại thất'] != '-', None)
df['Màu nội thất'] = df['Màu nội thất'].where(df['Màu nội thất'] != '-', None)
```

```python
provinces = pd.read_csv('Vietnam_province_info.csv')['Province/State']
```

```python
def extract(value):
    global provinces
    
    for province in provinces:
        if province in value:
            return province
    
    if "HCMC" in value:
        return "Ho Chi Minh City"
    elif "Peace" in value:
        return "Hoa Binh"
    elif "can Tho poetry" in value:
        return "Can Tho"
    elif "HCM City" in value:
        return "Ho Chi Minh City"
    elif "Danang" in value:
        return "Da Nang"
    elif "can tho" in value:
        return "Can Tho"
    elif "Ba Ria" in value or "Vung Tau" in value:
        return "Ba Ria - Vung Tau"
    elif "BAC Ninh City" in value:
        return "Bac Ninh"
    elif "Hue" in value:
        return "Tỉnh Thừa Thiên-Huế"

df['Address'] = df['Address'].apply(extract)

df.rename(columns={'Address': 'Tỉnh'}, inplace=True)

seller2_df = pd.read_csv('seller.csv')[['Mã tin', 'Tên']]

df['Bán bởi'] = pd.merge(car_df, seller2_df, on='Mã tin')['Tên']

df.drop(columns=['ad_id', 'Name', 'Mã tin'], inplace=True)

df
```

```python
df.to_csv('car_sell_processed.csv', index=False)
```
